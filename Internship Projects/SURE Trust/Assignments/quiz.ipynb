```python
def max(l):
  max1=l[0]
  for i in l:
    if i > max1:
      max1=i
  return max1
def min(l):
  min1=l[0]
  for i in l:
    if i < min1:
      min1=i
  return min1
l=[]
n=int(input("Enter size of list:"))
for i in range(n):
  print("Enter item for list:",i+1)
  item=int(input())
  l.append(item)
maximum=max(l)
minimum=min(l)
print("MAx =",maximum,"Mini =",minimum)
```

```python
def odd(a,b):
  l=[]
  for i in range(b,a-1,-1):
    if i%2 !=0:
      l.append(i)
  return l
a=int(input("Enter lower limit"))
b=int(input("Enter upper limit"))
if a<b:
  res=odd(a,b)
  print(res)
```
